# Alucar - Analisando as vendas

```python
import pandas as pd
```

```python
pd.read_csv("alucar.csv").head()
```

```python
alucar = pd.read_csv("alucar.csv")
```

```python
print(f"Quantidade de linhas e colunas {alucar.shape}")
```

```python
print(f"Quantidade de dados nulos: {alucar.isna().sum().sum()}" )
```

```python
alucar.dtypes
```

```python
alucar["mes"] = pd.to_datetime(alucar["mes"])
alucar.dtypes
```

```python
import seaborn as sns
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
print(sns.__version__)
```

```python
sns.lineplot(x="mes", y="vendas", data=alucar)
```

```python
sns.set_palette("Accent")
sns.set_style("darkgrid")
ax = sns.lineplot(x="mes", y="vendas", data=alucar)
ax.figure.set_size_inches(12, 6)
ax.set_title("Vendas Alucar de 2017 e 2018", loc="left", fontsize=18)
ax.set_xlabel("Tempo", fontsize=14)
ax.set_ylabel("Vendas", fontsize=14)
ax = ax
```

```python
alucar["aumento"] = alucar["vendas"].diff()
```

```python
alucar.head()
```

```python
sns.set_palette("Accent")
sns.set_style("darkgrid")
ax = sns.lineplot(x="mes", y="aumento", data=alucar)
ax.figure.set_size_inches(12, 6)
ax.set_title("Aumendo das vendas da Alucar de 2017 e 2018", loc="left", fontsize=18)
ax.set_xlabel("Tempo", fontsize=14)
ax.set_ylabel("Aumento", fontsize=14)
ax = ax
```

```python
def plotar(titulo, labelx, labely, x, y, dataset):
    sns.set_palette("Accent")
    sns.set_style("darkgrid")
    ax = sns.lineplot(x=x, y=y, data=dataset)
    ax.figure.set_size_inches(12, 6)
    ax.set_title(titulo, loc="left", fontsize=18)
    ax.set_xlabel(labelx, fontsize=14)
    ax.set_ylabel(labely, fontsize=14)
    ax = ax
```

```python
plotar("Aumento das vendas da Alucar de 2017 e 2018", "Tempo", "Aumento",
      "mes", "aumento", alucar)
```

```python
alucar["aceleracao"] = alucar["aumento"].diff()
```

```python
alucar.head()
```

```python
plotar("Acelerao das vendas da Alucar de 2017 e 2018", "Tempo", 
       "Acelerao", "mes", "aceleracao", alucar)
```

```python
plt.figure(figsize=(16, 12))
ax = plt.subplot(3,1,1)
ax.set_title("Anlise de vendas da Alucar de 2017 e 2018", fontsize=18, loc="left")
ax = sns.lineplot(x="mes", y="vendas", data=alucar)
plt.subplot(3,1,2)
ax = sns.lineplot(x="mes", y="aumento", data=alucar)
plt.subplot(3,1,3)
ax = sns.lineplot(x="mes", y="aceleracao", data=alucar)
```

```python
def plot_comparacao(x, y1, y2, y3, dataset, titulo):
    plt.figure(figsize=(16, 12))
    ax = plt.subplot(3,1,1)
    ax.set_title(titulo, fontsize=18, loc="left")
    ax = sns.lineplot(x=x, y=y1, data=dataset)
    plt.subplot(3,1,2)
    ax = sns.lineplot(x=x, y=y2, data=dataset)
    plt.subplot(3,1,3)
    ax = sns.lineplot(x=x, y=y3, data=dataset)
```

```python
plot_comparacao("mes", "vendas", "aumento", 'aceleracao', alucar,
               "Anlise das vendas da Alucar de 2017 e 2018")
```

```python
from pandas.plotting import autocorrelation_plot
```

```python
ax = plt.figure(figsize=(12,6))
ax.suptitle("Correlao das vendas", fontsize=18, x=0.26, y=0.95)
autocorrelation_plot(alucar["vendas"])
ax = ax
```

```python
ax = plt.figure(figsize=(12,6))
ax.suptitle("Correlao do aumento", fontsize=18, x=0.26, y=0.95)
autocorrelation_plot(alucar["aumento"][1:])
ax = ax
```

```python
ax = plt.figure(figsize=(12,6))
ax.suptitle("Correlao da acelerao", fontsize=18, x=0.26, y=0.95)
autocorrelation_plot(alucar["aceleracao"][2:])
ax = ax
```

```python
vendas_por_dia["aumento"] = vendas_por_dia["vendas"].diff()
vendas_por_dia["aceleracao"] = vendas_por_dia["aumento"].diff()
```

```python
plot_comparacao("dia", "vendas", "aumento", "aceleracao",
               vendas_por_dia, 
                "Anlise de vendas de Outubro e Novembro - Chocolura")
```

**Analisando a sazonalidade**

```python
vendas_por_dia["dia_da_semana"] = vendas_por_dia["dia"].dt.day_name()
```

```python
vendas_por_dia.head(7)
```

```python
vendas_por_dia["dia_da_semana"].unique()
```

```python
dias_traduzidos = {
    'Monday': "Segunda",
    'Tuesday': "Terca",
    'Wednesday': "Quarta",
    'Thursday': "Quinta",
    'Friday': "Sexta",
    'Saturday': "Sbado",
    'Sunday': "Domingo"
}
```

```python
vendas_por_dia["dia_da_semana"] = vendas_por_dia["dia_da_semana"].map(
    dias_traduzidos)
```

```python
vendas_por_dia.head(20)
```

# Agrupando os dias

```python
vendas_agrupadas = vendas_por_dia.groupby(
    "dia_da_semana")[["vendas", "aumento", "aceleracao"]].mean().round()
```
